```python
import pandas as pd
```

```python
df = pd.read_excel("movies.csv.xlsx")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['features'] = (
    df['Genre'] + ' ' +
    df['Director'] + ' ' +
    df['LeadActor'] + ' ' +
    df['Language']
)

df.head()
```

```python
# df[["Title", "combine_features"]].tail()
df['features'] = df['features'].str.replace('|', ' ', regex=False)
df['features'] = df['features'].str.lower()

df[['Title', 'features']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer()
feature_matrix = cv.fit_transform(df['features'])
similarity = cosine_similarity(feature_matrix)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()

    if movie_name not in df['Title'].str.lower().values:
        print("Movie not found in dataset")
        return
    index = df[df['Title'].str.lower() == movie_name].index[0]
    distances = list(enumerate(similarity[index]))
    movies_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]
    print("Movies recommended for:", df.iloc[index]['Title'])
    print("--------------------------------")
    for i in movies_list:
        print(df.iloc[i[0]]['Title'])
movie_name = input("Enter the movie name: ");
recommend(movie_name)
```
